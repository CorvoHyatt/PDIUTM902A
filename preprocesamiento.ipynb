# Clasificacion de imagenes

```python
import os
import shutil

def clasificar_imagenes(origen, destino_leaf, destino_early):
    # Verificar si las carpetas de destino existen, si no, crearlas
    if not os.path.exists(destino_leaf):
        os.makedirs(destino_leaf)
    if not os.path.exists(destino_early):
        os.makedirs(destino_early)

    # Listar archivos en la carpeta de origen
    archivos = os.listdir(origen)

    for archivo in archivos:
        ruta_completa = os.path.join(origen, archivo)

        # Verificar si el nombre del archivo contiene "leaf" o "Early"
        if "Leaf" in archivo:
            shutil.move(ruta_completa, os.path.join(destino_leaf, archivo))
            print(f"Imagen {archivo} movida a {destino_leaf}")
        elif "Early" in archivo:
            shutil.move(ruta_completa, os.path.join(destino_early, archivo))
            print(f"Imagen {archivo} movida a {destino_early}")

# Rutas de las carpetas de origen y destino
carpeta_origen = r"C:\Users\Rv_Ru\OneDrive\Documentos\UTM\9no\pdi\images"
carpeta_destino_leaf = r"C:\Users\Rv_Ru\OneDrive\Documentos\UTM\9no\pdi\imagenes\leaf_mold"
carpeta_destino_early = r"C:\Users\Rv_Ru\OneDrive\Documentos\UTM\9no\pdi\images\early_blight"

# Llamar a la función para clasificar las imágenes
clasificar_imagenes(carpeta_origen, carpeta_destino_leaf, carpeta_destino_early)
```

# Renombrar imagenes

```python
import os

def renombrar_imagenes(carpeta_origen, nuevo_nombre):
    # Listar archivos en la carpeta de origen
    archivos = os.listdir(carpeta_origen)

    # Inicializar contador
    contador = 1

    for archivo in archivos:
        # Obtener la extensión del archivo
        nombre, extension = os.path.splitext(archivo)

        # Crear el nuevo nombre con el formato especificado
        nuevo_nombre_archivo = f"{nuevo_nombre}_{contador}{extension}"

        # Crear la ruta completa de los archivos antiguo y nuevo
        ruta_antigua = os.path.join(carpeta_origen, archivo)
        ruta_nueva = os.path.join(carpeta_origen, nuevo_nombre_archivo)

        # Renombrar el archivo
        os.rename(ruta_antigua, ruta_nueva)
        
        print(f"Imagen {archivo} renombrada a {nuevo_nombre_archivo}")

        # Incrementar el contador
        contador += 1

# Rutas de la carpeta de origen y nuevo nombre
carpeta_origen = r"C:\Users\Rv_Ru\OneDrive\Documentos\UTM\9no\pdi\imagenes\early_blight"
nuevo_nombre = "early_blight"

# Llamar a la función para renombrar las imágenes
renombrar_imagenes(carpeta_origen, nuevo_nombre)
```
